```python
# pd.set_option() to set the number of characters, a column can occupy
import pandas as pd
pd.set_option('display.max_colwidth',100)
dataset = pd.read_csv("SMSSpamCollection.txt", sep = '\t', header = None)
dataset.columns = ['label', 'body_text']
dataset.head()
```

```python
# Removing Punctuations from text
# join clubs the char together over space or something. In this case it is "" as we have a list of characters including space in it.
# List comprehension to classify punct from non-punct
# Lambda() to apply the remove_punct() to every row in our data set.

import string
def remove_punct(text):
    text_no_punct = "".join([char for char in text if char not in string.punctuation])
    return text_no_punct
dataset['body_text_no_punct'] = dataset['body_text'].apply(lambda x : remove_punct(x))
dataset.head()
    
```

```python
# Tokenizing the text
# '\W+' is the re for splitting text at one or more non-word character
import re
def tokenize(text):
    tokens = re.split('\W+',text)
    return tokens
dataset['body_text_tokenized'] = dataset['body_text_no_punct'].apply(lambda x : tokenize(x.lower()))
dataset.head()
```

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopwords] 
    return text
dataset['body_text_cleaned'] = dataset['body_text_tokenized'].apply(lambda x: remove_stopwords(x))
dataset.head()
```
